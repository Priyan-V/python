```python
import numpy as np
a=np.array([[1,2,5],[2,3,1],[5,6,8]])
b=np.array([[2,8,7],[6,9,8],[5,6,8]])
print("The first matrix:")
print(a)
print("\n The second matrix:")
print(b)
add=np.add(a,b)
print("The sum:")
print(add)
sub=np.subtract(a,b)
print("The difference:")
print(sub)
mul=np.dot(a,b)
print("The product:")
print(mul)
transposeA=np.transpose(a)
print("The transpose of first matrix:")
print(transposeA)
transposeB=np.transpose(b)
print("The transpose of second matrix:")
print(transposeB)
```

```python
import matplotlib.pyplot as plt
x=[1,2,5,2,3,1,5,6,8]
y=[2,8,7,6,9,8,5,6,8]
plt.plot(x,y)
plt.title("Normal graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()

plt.figure()
plt.stem(x,y)
plt.title("Stem graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

plt.figure()
plt.bar(x,y)
plt.title("Bar graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
